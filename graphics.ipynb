```python
with open("results.txt", "r") as f:
    time_line = int(f.readline())
    row, col, block = [], [], [] 
    for line in f:
        buff = line.split()
        row.append(int(buff[0]))
        col.append(int(buff[1]))
        block.append(int(buff[2]))
```

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

x = [2**i for i in range(len(row))]

plt.plot(x, row, label = "разбиение по строкам")
plt.plot(x, col, label = "разбиение по столбцам")
plt.plot(x, block, label = "разбиение по блокам")
plt.xlabel("Количество потоков")
plt.ylabel("Время выполнения")
plt.legend()
plt.show()
```

```python
row_s = [time_line/i for i in row] 
col_s = [time_line/i for i in col]
block_s = [time_line/i for i in block]
row_ef = [time_line/row[i]/x[i] for i in range(len(row))]
col_ef = [time_line/col[i]/x[i] for i in range(len(row))]
block_ef = [time_line/block[i]/x[i] for i in range(len(row))]
```

```python
plt.plot(x, row_s, label = "разбиение по строкам")
plt.plot(x, col_s, label = "разбиение по столбцам")
plt.plot(x, block_s, label = "разбиение по блокам")
plt.xlabel("Количество потоков")
plt.ylabel("Ускорение алгоритмов")
plt.legend()
plt.show()
```

```python
plt.plot(x, row_ef, label = "разбиение по строкам")
plt.plot(x, col_ef, label = "разбиение по столбцам")
plt.plot(x, block_ef, label = "разбиение по блокам")
plt.xlabel("Количество потоков")
plt.ylabel("Эффективность алгоритмов")
plt.legend()
plt.show()
```
