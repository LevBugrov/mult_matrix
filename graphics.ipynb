```python
with open("results.txt", "r") as f:
    time_line = int(f.readline())
    row, col, block = [], [], [] 
    for line in f:
        buff = line.split()
        row.append(buff[0])
        col.append(buff[1])
        block.append(buff[2])
```

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

x = [2**i for i in range(len(row))]

plt.plot(x, row, label = "line 1")
plt.plot(x, col, label = "line 2")
plt.plot(x, block, label = "line 3")

plt.title("Время работы алгоритмов")

plt.xlabel("Количество потоков")
plt.ylabel("Время выполнения")


plt.legend()

plt.show()
```

```python
row_s = [time_line/i for i in row] 
col_s = [time_line/i for i in col]
block_s = [time_line/i for i in block]
row_ef = [time_line/row[i]/x[i] for i in range(len(row))]
col_ef = [time_line/col[i]/x[i] for i in range(len(row))]
block_ef = [time_line/block[i]/x[i] for i in range(len(row))]
```
